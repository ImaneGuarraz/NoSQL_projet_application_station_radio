```python
import pandas as pd
```

```python
# charger le dataset
df = pd.read_csv("data.csv")
```

```python
# Renommer les colonnes pour avoir des noms propres
df = df.rename(columns={
    "StationName": "station_name",
    "StationID": "station_id",
    "Country": "country",
    "City": "city",
    "Genre": "genre",
    "StreamURL": "stream_url",
    "TrackTitle": "track_title",
    "TrackID": "track_id",
    "ArtistName": "artist_name",
    "DurationSec": "duration",
    "PlayTimestamp": "play_timestamp",
    "Listeners": "listener_count"
})
```

```python
# Nettoyer les valeurs manquantes
df["Genre"] = df["Genre"].fillna("Unknown")
df["Country"] = df["Country"].fillna("Unknown")
df["City"] = df["City"].fillna("Unknown")
```

```python
# Supprimer les lignes complètement vides ou inutilisables
df = df.dropna(subset=["station_id", "station_name", "stream_url"])
```

```python
# Convertir les types
df["duration"] = pd.to_numeric(df["duration"], errors="coerce")
df["play_timestamp"] = pd.to_datetime(df["play_timestamp"], errors="coerce")
```

```python
# Supprimer les doublons
df = df.drop_duplicates()
```

```python
# Normaliser les genres
df["genre"] = df["genre"].str.strip().str.lower()
```

```python
# Sauvegarder le dataset nettoyé
df.to_csv("data_clean.csv", index=False)
```
